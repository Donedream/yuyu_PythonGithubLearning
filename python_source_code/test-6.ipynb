```python
import matplotlib as plt
import matplotlib.pyplot as plt2
```

示例·一个月内每天的销售量

```python
days = list(range(1, 31))
sales = [250, 300, 280, 310, 290, 330, 340, 360, 300, 310,
         280, 350, 360, 390, 370, 400, 420, 430, 410, 390,
         400, 420 ,430, 440, 450, 460, 480, 500, 490 ,520]
```

```python
fig = plt2.figure(figsize = (10, 6))
#plt2.show() # 仅设定figure的尺寸最后无法生成图像出来
```

```python
plt2.bar(days, sales, color = 'lightblue', alpha = 0.7, label = 'Daliy Sales')
```

```python
plt2.plot(days, sales, color = 'blue', marker = 'o', linestyle = '-', label = 'Trend')
```

```python
plt2.bar(days, sales, color = 'lightblue', alpha = 0.7, label = 'Daliy Sales') # 这里表示绘制图像的代码段必须在一起执行·不然相当于在空白的图像上重新绘制·而且要注意横纵坐标的数值也会发生变化
plt2.title('Daily Sales for a Month')
plt2.xlabel('Day')
plt2.ylabel('Sales')
plt2.legend() # 显示如图所示的左上角的图例/label的作用
```

```python
import plotly.graph_objects as go
```

```python
bar = go.Bar(
    x = days,
    y = sales,
    name = 'Daily Sales',
    marker = dict(color = 'lightblue'),
    opacity = 0.7
)
```

```python
line = go.Scatter(
    x = days,
    y = sales,
    name = 'Trend',
    mode = 'lines+markers',
    marker = dict(color = 'blue')
)
```

```python
fig2 = go.Figure(data = [bar, line])
fig2.show()
```

```python
fig2.update_layout(
    title = 'Daily Sales for a Month',
    xaxis_title = 'Day',
    yaxis_title = 'Sales',
    legend_title = 'Legend' # 好像默认需要多个图像类型才会显示图例
)
```
